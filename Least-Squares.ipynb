```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Function to Read Data

def read_data(name):

    # Read in data
    x,y = [],[]
    f = open(name,'r')  
    header_flag = 0
    for line in f:
        #Doesn't read header of the file
        if header_flag == 0: 
            header_flag = 1
            continue
        else:
            fields = line.rstrip('\n').split()  
            x += [float(fields[0])]                       
            y += [float(fields[1])]
 
    f.close()
    return x,y

#Plotting Functions

def plot_scatter(x,y,para,A,model_number):
    
    plt.figure()
    plt.scatter(x,y,alpha = 0.5,label = 'data')
    plt.scatter(x,np.dot(A(x),para),color = 'black',alpha = 0.7,label = 'best_fit')
    plt.title('lsq_fit_model_{}'.format(model_number))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()
    
    return

def plot_line(x,y,para,A,model_number):
    
    plt.figure()
    plt.scatter(x,y,alpha = 0.5,label = 'data')
    plt.plot(np.sort(x),np.dot(A(np.sort(x)),para),color = 'black',alpha = 1.0,label = 'best_fit',linewidth=2.5)
    plt.title('lsq_fit_model_{}'.format(model_number))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()
    
    return
```

```python
#Pseudo inverse = ((A^T*A)^-1*A^T)
lsqt_fit = lambda x,y,A : \
                np.matmul(np.matmul(np.linalg.inv(np.matmul(np.transpose(A(x)),A(x))),np.transpose(A(x))),y)

#Funtion to evaluate MSE
mse = lambda x,y,params,A : (1/len(x))*np.sum((np.power(np.array(np.dot(A(x),params))-np.array(y),2)))
```

```python
#Defining the 3 models

A1 = lambda x : np.transpose([np.ones(len(x)),np.power(x,1.0),np.power(x,2.0),np.power(x,3.0)])
A2 = lambda x : np.transpose([np.ones(len(x)),np.exp(x),np.exp(2*x),np.exp(3*x)])
A3 = lambda x : np.transpose([np.sin(x),np.cos(x),np.sin(2*x),np.cos(2*x)])
```

```python
#Driver Code 

x,y = read_data('Data_files/data-1.txt')
x,y = np.array(x),np.array(y)

para_1,para_2,para_3 = lsqt_fit(x,y,A1),lsqt_fit(x,y,A2),lsqt_fit(x,y,A3)

print('optimal_parameter for model 1 is {}'.format(para_1))
plot_line(x,y,para_1,A1,1)
print('MSE for model 1 is {}\n'.format(mse(x,y,para_1,A1)))

print('optimal_parameter for model 2 is {}'.format(para_2))
plot_line(x,y,para_2,A2,2)
print('MSE for model 2 is {}\n'.format(mse(x,y,para_2,A2)))

print('optimal_parameter for model 3 is {}'.format(para_3))
plot_line(x,y,para_3,A3,3)
print('MSE for model 3 is {}\n'.format(mse(x,y,para_3,A3)))
```
