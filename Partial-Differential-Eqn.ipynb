### 6. PDE: Solving the Heat Equation 

Heat Equation: 

$$ \frac{d^{2}T}{dx^{2}} - \frac{dT}{dt} = 0; T(0,t) = 0; T(L,t) = 0; T(x,0) = sin(\pi x) $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.1                          #no of spatial grids
k = 0.025                        #no of time grids
x = np.arange(0,1.0+h,h)
t = np.arange(0,0.25+k,k)

T = np.zeros([len(t),len(x)])   #Initial solution
T[0,:] = np.sin(np.pi*x)        #Initial condition
T[0,0],T[0,-1] = 0.0,0.0        #Boundary condition

#Explicit Method
for i in range(1,len(t)):
    for j in range(1,len(x)-1):
        T[i,j] = (k/(h**2))*T[i-1,j+1] + (1 - 2*(k/(h**2)))*T[i-1,j] + (k/(h**2))*T[i-1,j-1]

#Plotting
plt.figure()

for count_i,i in enumerate(T):
    plt.plot(x,i,label = '{} s'.format(np.round(t[count_i],3)))
    plt.scatter(x,i)

plt.legend(bbox_to_anchor=[1.25, 1.0]) 
plt.ylabel('Temperature')
plt.xlabel('X')
plt.show()
```
